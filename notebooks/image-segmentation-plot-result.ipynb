```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_resunetplusplus = pd.read_csv('../input/kvasit-seg-train-test-valid/model_comparison_files/model_comparision_files/renunetplusplus_8.csv',encoding="utf8")
data_resunet = pd.read_csv('../input/kvasit-seg-train-test-valid/model_comparison_files/model_comparision_files/resunet_8.csv',encoding="utf8")
data_unet = pd.read_csv('../input/kvasit-seg-train-test-valid/model_comparison_files/model_comparision_files/unet_8.csv',encoding="utf8")
```

```python
def plot_curve(df, training_metric, validation_metric):
    plt.plot(df[training_metric], label='{} (training data)'.format(training_metric))
    plt.plot(df[validation_metric], label='{} (validation data)'.format(training_metric))
    plt.title('{} FOR TRAINING AND VALIDATION DATA'.format(training_metric))
    plt.ylabel('{} value'.format(training_metric))
    plt.xlabel('No. epoch')
    plt.legend(loc="upper left")
    plt.show()
```

```python
data_unet.columns
```

```python
plot_curve(data_unet, 'loss', 'val_loss')
plot_curve(data_unet, 'precision_1', 'val_precision_1')
plot_curve(data_unet, 'recall_1', 'val_recall_1')
plot_curve(data_unet, 'dice_coef', 'val_dice_coef')
plot_curve(data_unet, 'mean_io_u_1', 'val_mean_io_u_1')
```
